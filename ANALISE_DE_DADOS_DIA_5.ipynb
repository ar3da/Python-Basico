```python
import pandas as pd
import matplotlib.pyplot as plt


#Exemplos de notas de alunos

dados = {
    "Alunos": ["Ana", "Bruno", "Daniela", "Eduardo", "Fernanda"],
    "Notas": [8, 7.5, 9, 6, 7]
}

df= pd.DataFrame(dados)

print(df)
```

```python
#Criando um gr√°fico de barras

plt.bar(df["Alunos"], df["Notas"])
plt.xlabel("Alunos")
plt.ylabel("Notas")
plt.title("Notas dos Alunos")
plt.show()
```

```python
#Criando gr√°ficos de linhas

plt.plot(df["Alunos"], df["Notas"])
plt.xlabel("Alunos")
plt.ylabel("Notas")
plt.title("Notas dos Alunos")
plt.show()
```

```python
#Criando gr√°fico de pizza

plt.pie(df["Notas"], labels=df["Alunos"])
plt.title("Notas dos Alunos")
plt.show()
```

```python
#Exercicio dia 5: Criar gr√°ficos usando dados de vendas

dados = {
    'Produto': ['Caderno', 'Caneta', 'L√°pis', 'Borracha'],
    'Vendas': [120, 200, 150, 90]
}

df = pd.DataFrame(dados)

print(df)

plt.bar(df["Produto"], df["Vendas"])
plt.xlabel("Produto")
plt.ylabel("Vendas")
plt.title("Vendas dos Produtos")
plt.show()

print("_"*40)

plt.plot(df["Produto"], df["Vendas"])
plt.xlabel("Produto")
plt.ylabel("Vendas")
plt.title("Vendas dos Produtos")
plt.show()

print("_"*40)

plt.pie(df["Vendas"], labels=df["Produto"])
plt.title("Vendas dos Produtos")
plt.show()
```
