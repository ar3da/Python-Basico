```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Criando um DataFrame de exemplo
dados = {
    "Idade": [23, 25, 31, 22, 28, 35, 40, 29, 30, 27],
    "Salario": [2000, 2500, 4000, 1800, 3000, 5000, 7000, 3200, 3500, 2800],
    "Departamento": ["TI", "RH", "TI", "RH", "Financeiro", "TI", "Financeiro", "RH", "TI", "Financeiro"]
}

df = pd.DataFrame(dados)
print(df)

print("_"*40)

# --- Histograma ---

plt.figure(figsize=(6,4))
plt.hist(df["Idade"], bins=5, color="skyblue", edgecolor="black")
plt.title("Distribuição de Idades")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()

print("_"*40)

# --- Boxplot ---
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Salario"], color="lightgreen")
plt.title("Distribuição de Salários (Boxplot)")
plt.show()

print("_"*40)

# --- Scatterplot ---
plt.figure(figsize=(6,4))
sns.scatterplot(x="Idade", y="Salario", data=df, hue="Departamento", palette="Set2", s=100)
plt.title("Idade x Salário por Departamento")
plt.show()

print("_"*40)

# --- Heatmap de correlação ---
plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="Blues")
plt.title("Correlação entre Variáveis")
plt.show()
```
